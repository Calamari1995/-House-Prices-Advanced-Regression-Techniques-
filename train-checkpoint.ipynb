```python
train = Processor(data='train.csv', with_pandas=True)
#test = Processor(data='test.csv', with_pandas=True)
target = 'SalePrice'
y_data = np.log(train.data[target].copy())
#y_data /= 1e5

train._DROP = {
    
    1 : [
        
        'Id',
        'YrSold',
        target
    ]
    
}

train.data = train._drop(train.data, 1)

train._FILLNA = {
    
    '<unk>' : train.data.select_dtypes(include='object').columns.to_list(),
    0 : train.data.select_dtypes(exclude='object').columns.to_list(),
    
    
}

train._CATEGORICAL = train.data.select_dtypes(include='object').columns.to_list()

train.data = train._fillna(train.data)
train.data = train._to_categorical(train.data)
```

```python
import seaborn as sns

sns.kdeplot(np.log(train.data.SalePrice))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train.data, y_data, random_state=8, test_size=0.1)
```

```python
model1 = xgboost.XGBRegressor(n_estimators=50, max_depth=3).fit(X_train, y_train)
```
